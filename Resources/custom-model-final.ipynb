```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (confusion_matrix, classification_report,
                             precision_score, recall_score, f1_score,
                             ConfusionMatrixDisplay)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import (GlobalAveragePooling2D, Dense, Dropout,
                                     BatchNormalization, Conv2D, MaxPooling2D,
                                     Input, Activation)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, Callback
from tensorflow.keras.regularizers import l2

# Set seeds
np.random.seed(42)
tf.random.set_seed(42)

print("="*70)
print("üöÄ SUPERVISED LEARNING FOR MANGO LEAF DISEASE")
print("="*70)

# ============================================================================
# CONFIGURATION
# ============================================================================
DATA_DIR = '/kaggle/input/mango-leaf-disease-dataset'
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

# Training configuration
TRAIN_CONFIG = {
    'epochs': 100,  # More epochs for supervised learning
    'early_stopping_patience': 20,
    'lr_patience': 8,
    'initial_lr': 1e-3,
}

# ============================================================================
# LOAD DATA
# ============================================================================
print("\n=== Loading Dataset ===")
filepaths = []
labels = []
for label in sorted(os.listdir(DATA_DIR)):
    label_path = os.path.join(DATA_DIR, label)
    if os.path.isdir(label_path):
        for file in os.listdir(label_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                filepaths.append(os.path.join(label_path, file))
                labels.append(label)

df = pd.DataFrame({'filename': filepaths, 'class': labels})
classes = sorted(df['class'].unique())
num_classes = len(classes)

print(f"Total samples: {len(df)}")
print(f"Classes: {classes}")

# Display class distribution
print("\nClass distribution:")
class_counts = df['class'].value_counts()
for cls, count in class_counts.items():
    print(f"  {cls}: {count} samples ({100*count/len(df):.1f}%)")

# ============================================================================
# SPLIT DATA
# ============================================================================
print("\n=== Splitting Data ===")
# First split: separate test set
train_val_df, test_df = train_test_split(
    df, test_size=0.1, stratify=df['class'], random_state=SEED
)

# Second split: separate training and validation
train_df, val_df = train_test_split(
    train_val_df, test_size=0.111, stratify=train_val_df['class'], random_state=SEED
)

print(f"Data split:")
print(f"  Training: {len(train_df)} ({100*len(train_df)/len(df):.1f}%)")
print(f"  Validation: {len(val_df)} ({100*len(val_df)/len(df):.1f}%)")
print(f"  Test: {len(test_df)} ({100*len(test_df)/len(df):.1f}%)")

# ============================================================================
# DATA GENERATORS WITH AGGRESSIVE AUGMENTATION
# ============================================================================
print("\n=== Creating Data Generators ===")

# Aggressive augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.25,
    height_shift_range=0.25,
    shear_range=0.2,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.7, 1.3],
    fill_mode='nearest',
    channel_shift_range=20.0,
)

# Only rescaling for validation and test
val_test_datagen = ImageDataGenerator(rescale=1./255)

class_names = sorted(df['class'].unique())

# ============================================================================
# BUILD REGULARIZED MODEL
# ============================================================================
def build_model(input_shape=(224, 224, 3), num_classes=8):
    """
    Model with heavy regularization to prevent overfitting
    """
    inputs = Input(shape=input_shape)
    
    # Block 1
    x = Conv2D(32, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.2)(x)
    
    # Block 2
    x = Conv2D(64, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    
    # Block 3
    x = Conv2D(128, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.4)(x)
    
    # Block 4
    x = Conv2D(256, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.5)(x)
    
    # Head with heavy dropout
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.6)(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

print("\n=== Building Model ===")
model = build_model(input_shape=IMG_SIZE + (3,), num_classes=num_classes)
print(f"Model parameters: {model.count_params():,}")

# ============================================================================
# CUSTOM CALLBACK
# ============================================================================
class TrainingMonitor(Callback):
    def __init__(self):
        super().__init__()
        self.best_val_acc = 0
        
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        val_acc = logs.get('val_accuracy', 0)
        if val_acc > self.best_val_acc:
            self.best_val_acc = val_acc
            print(f"\nüéØ New best val_acc: {val_acc:.4f}")

# ============================================================================
# CREATE DATA GENERATORS
# ============================================================================
print("\n=== Preparing Data Generators ===")

train_gen = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=SEED,
    classes=class_names
)

val_gen = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    classes=class_names
)

test_gen = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    classes=class_names
)

# ============================================================================
# CALCULATE CLASS WEIGHTS (for imbalanced datasets)
# ============================================================================
print("\n=== Calculating Class Weights ===")
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(train_df['class']),
    y=train_df['class']
)
class_weights_dict = {i: weight for i, weight in enumerate(class_weights)}

print("Class weights:")
for i, (cls, weight) in enumerate(zip(class_names, class_weights)):
    print(f"  {cls}: {weight:.3f}")

# ============================================================================
# COMPILE AND TRAIN MODEL
# ============================================================================
print("\n" + "="*70)
print("TRAINING MODEL")
print("="*70)

# Compile model
model.compile(
    optimizer=Adam(learning_rate=TRAIN_CONFIG['initial_lr']),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=TRAIN_CONFIG['early_stopping_patience'],
    restore_best_weights=True,
    mode='max',
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=TRAIN_CONFIG['lr_patience'],
    min_lr=1e-7,
    mode='max',
    verbose=1
)

monitor = TrainingMonitor()

# Train model
print("\nTraining model with all labeled data...")
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=TRAIN_CONFIG['epochs'],
    class_weight=class_weights_dict,  # Handle class imbalance
    callbacks=[early_stop, reduce_lr, monitor],
    verbose=2
)

# ============================================================================
# EVALUATE MODEL
# ============================================================================
print("\n" + "="*70)
print("MODEL EVALUATION")
print("="*70)

# Evaluate on validation set
val_loss, val_acc = model.evaluate(val_gen, verbose=1)
print(f"\nüìä Validation Results:")
print(f"   Accuracy: {val_acc:.4f}")
print(f"   Loss: {val_loss:.4f}")

# Evaluate on test set
test_loss, test_acc = model.evaluate(test_gen, verbose=1)
print(f"\nüéØ Test Results:")
print(f"   Accuracy: {test_acc:.4f}")
print(f"   Loss: {test_loss:.4f}")

# ============================================================================
# DETAILED METRICS
# ============================================================================
print("\n" + "="*70)
print("DETAILED METRICS")
print("="*70)

# Get predictions
y_pred_probs = model.predict(test_gen, verbose=0)
y_pred_labels = np.argmax(y_pred_probs, axis=1)
y_true_labels = test_gen.classes

# Classification report
print("\nClassification Report:")
print(classification_report(y_true_labels, y_pred_labels, target_names=classes))

# Calculate metrics
test_precision = precision_score(y_true_labels, y_pred_labels, average='macro')
test_recall = recall_score(y_true_labels, y_pred_labels, average='macro')
test_f1 = f1_score(y_true_labels, y_pred_labels, average='macro')

print(f"\nüìä Summary Metrics:")
print(f"   Precision: {test_precision:.4f}")
print(f"   Recall: {test_recall:.4f}")
print(f"   F1-Score: {test_f1:.4f}")

# ============================================================================
# VISUALIZATION
# ============================================================================
print("\n=== Generating Visualizations ===")

# Training curves
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Accuracy plot
axes[0].plot(history.history['accuracy'], label='Training', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Validation', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Loss plot
axes[1].plot(history.history['loss'], label='Training', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Validation', linewidth=2)
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
plt.show()

# Confusion Matrix
cm = confusion_matrix(y_true_labels, y_pred_labels)
plt.figure(figsize=(12, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot(cmap='Blues', values_format='d', xticks_rotation=45)
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

# Per-class accuracy
class_accuracies = {}
for i, cls in enumerate(classes):
    mask = (y_true_labels == i)
    if mask.sum() > 0:
        class_acc = (y_pred_labels[mask] == y_true_labels[mask]).mean()
        class_accuracies[cls] = class_acc

print("\nPer-class Accuracy:")
for cls, acc in sorted(class_accuracies.items(), key=lambda x: x[1], reverse=True):
    print(f"  {cls}: {acc:.4f}")

print("\n‚úÖ Supervised learning completed!")
```

```python
# Add these imports to the top of your file
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from sklearn.manifold import TSNE
import matplotlib.cm as cm
from matplotlib.colors import ListedColormap
```

```python
# ============================================================================
# VISUALIZATION: GRAD-CAM & T-SNE
# ============================================================================

# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    """Creates a Grad-CAM heatmap for a given image and model."""
    grad_model = Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def create_gradcam_overlay(img_path, heatmap, alpha=0.4):
    """Creates the superimposed Grad-CAM image and returns it as an array."""
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
    img = tf.keras.preprocessing.image.img_to_array(img)
    heatmap = np.uint8(255 * heatmap)
    jet = cm.get_cmap("jet")
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]
    jet_heatmap = tf.keras.preprocessing.image.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = tf.keras.preprocessing.image.img_to_array(jet_heatmap)
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = tf.keras.preprocessing.image.array_to_img(superimposed_img)
    return superimposed_img

def create_feature_extractor(model, layer_name='global_average_pooling2d'):
    """Creates a new model that outputs features from a specified layer."""
    return Model(inputs=model.inputs, outputs=model.get_layer(layer_name).output)

# ============================================================================
# GENERATE GRAD-CAM VISUALIZATIONS (2x4 GRID FOR EACH CLASS)
# ============================================================================
print("\n" + "="*70)
print("GENERATING GRAD-CAM VISUALIZATIONS (ONE PER CLASS)")
print("="*70)

# Find the name of the last convolutional layer
last_conv_layer_name = None
for layer in reversed(model.layers):
    if isinstance(layer, tf.keras.layers.Conv2D):
        last_conv_layer_name = layer.name
        break

print(f"Using layer '{last_conv_layer_name}' for Grad-CAM")

# --- Find the best representative image for each class ---
print("Finding best representative images for each class...")
best_samples = []

for class_name in class_names:
    class_df = test_df[test_df['class'] == class_name]
    best_sample = None
    best_confidence = -1.0

    # Try to find a correctly predicted image with the highest confidence
    for _, row in class_df.iterrows():
        img_path = row['filename']
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array_normalized = np.expand_dims(img_array, axis=0) / 255.0
        
        preds = model.predict(img_array_normalized, verbose=0)
        pred_class_idx = np.argmax(preds[0])
        pred_class_name = class_names[pred_class_idx]
        confidence = preds[0][pred_class_idx]
        
        if pred_class_name == class_name and confidence > best_confidence:
            best_confidence = confidence
            best_sample = {
                'img_path': img_path,
                'img_array': img_array / 255.0,
                'confidence': confidence,
                'pred_class': pred_class_name
            }

    # Fallback: if no correct predictions, take the first image
    if best_sample is None:
        print(f"  Warning: No correct predictions found for class '{class_name}'. Using first available sample.")
        first_row = class_df.iloc[0]
        img_path = first_row['filename']
        img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        img_array_normalized = np.expand_dims(img_array, axis=0) / 255.0
        
        preds = model.predict(img_array_normalized, verbose=0)
        pred_class_idx = np.argmax(preds[0])
        pred_class_name = class_names[pred_class_idx]
        confidence = preds[0][pred_class_idx]

        best_sample = {
            'img_path': img_path,
            'img_array': img_array / 255.0,
            'confidence': confidence,
            'pred_class': pred_class_name
        }
    
    best_samples.append(best_sample)
    print(f"  Found sample for '{class_name}' with confidence {best_sample['confidence']:.2f}")

# --- Generate the 2x4 subplot ---
print("\nGenerating Grad-CAM grid plot...")
num_classes = len(class_names)
ncols = 4
nrows = (num_classes + ncols - 1) // ncols

fig, axes = plt.subplots(nrows, ncols, figsize=(20, 10))
axes = axes.flatten()

for i, sample_data in enumerate(best_samples):
    img_path = sample_data['img_path']
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array_normalized = np.expand_dims(img_array, axis=0) / 255.0
    
    pred_class_idx = class_names.index(sample_data['pred_class'])
    heatmap = make_gradcam_heatmap(img_array_normalized, model, last_conv_layer_name, pred_class_idx)
    gradcam_overlay_img = create_gradcam_overlay(img_path, heatmap)

    ax = axes[i]
    ax.imshow(gradcam_overlay_img)
    
    title = f"True: {class_names[i]}\nPred: {sample_data['pred_class']} ({sample_data['confidence']:.2f})"
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.axis('off')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.suptitle('Grad-CAM Visualization for Each Class', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('gradcam_grid_all_classes.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n‚úÖ Grad-CAM grid visualization completed!")


# ============================================================================
# GENERATE T-SNE VISUALIZATION
# ============================================================================
print("\n" + "="*70)
print("GENERATING T-SNE VISUALIZATION")
print("="*70)

# Create feature extractor
feature_extractor = create_feature_extractor(model)

# Extract features from the test set
print("Extracting features from test set...")
features = []
true_labels = []

# Process in batches to avoid memory issues
batch_size = 32
num_batches = int(np.ceil(len(test_df) / batch_size))

for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(test_df))
    batch_df = test_df.iloc[start_idx:end_idx]
    
    batch_images = []
    batch_labels = []
    
    for _, row in batch_df.iterrows():
        img = tf.keras.preprocessing.image.load_img(row['filename'], target_size=IMG_SIZE)
        img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
        batch_images.append(img_array)
        batch_labels.append(row['class'])
    
    batch_features = feature_extractor.predict(np.array(batch_images), verbose=0)
    features.extend(batch_features)
    true_labels.extend(batch_labels)
    
    if (i + 1) % 10 == 0:
        print(f"Processed {i + 1}/{num_batches} batches")

features = np.array(features)
print(f"Extracted features shape: {features.shape}")

# Apply t-SNE
print("Applying t-SNE...")
tsne = TSNE(n_components=2, random_state=SEED, perplexity=30, n_iter=1000, verbose=1)
tsne_results = tsne.fit_transform(features)

# Plot t-SNE results
plt.figure(figsize=(12, 10))
colors = plt.cm.tab10(np.linspace(0, 1, len(class_names)))

for i, cls in enumerate(class_names):
    indices = [j for j, label in enumerate(true_labels) if label == cls]
    plt.scatter(
        tsne_results[indices, 0], 
        tsne_results[indices, 1],
        c=[colors[i]],
        label=cls,
        alpha=0.7
    )

plt.title('t-SNE Visualization of Learned Features', fontsize=16, fontweight='bold')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('tsne_visualization.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n‚úÖ All visualizations completed!")
```
