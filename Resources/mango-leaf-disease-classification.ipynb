```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import random
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
import keras
from keras.models import Sequential
from keras.optimizers import Adam, Adamax
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.layers import MaxPooling2D, Flatten, Dense, BatchNormalization, GlobalAveragePooling2D, Conv2D, Dropout, Rescaling, Input
from keras import regularizers
from keras.callbacks import EarlyStopping, ModelCheckpoint

# Configuration
data_path = "/kaggle/input/mango-leaf-disease-dataset"
img_size = (224, 224)
batch_size = 64
mode = "rgb"
epochs = 100
SEED = 42

# Create dataframes
def df_maker(path):
    file_paths = []
    labels = []

    folds = os.listdir(path)
    for fold in folds:
        fold_path = os.path.join(path, fold)
        file_list = os.listdir(fold_path)
        for file in file_list:
            file_path = os.path.join(fold_path, file)
            file_paths.append(file_path)
            labels.append(fold)

    file_series = pd.Series(file_paths, name="file_paths")
    label_series = pd.Series(labels, name="labels")

    df = pd.concat([file_series, label_series], axis=1)
    return df

# Create dataframe
df = df_maker(data_path)

# Split data
train_df, test_df = train_test_split(df, train_size=0.8, shuffle=True, random_state=SEED, stratify=df["labels"])
test_df, val_df = train_test_split(test_df, train_size=0.5, shuffle=True, random_state=SEED, stratify=test_df["labels"])

# Data generators
train_datagen = ImageDataGenerator()
test_datagen = ImageDataGenerator()
val_datagen = ImageDataGenerator()

train_data = train_datagen.flow_from_dataframe(
    train_df,
    x_col="file_paths",
    y_col="labels",
    target_size=img_size,
    color_mode=mode,
    class_mode='categorical',
    batch_size=batch_size,
    shuffle=True,
)
test_data = test_datagen.flow_from_dataframe(
    test_df,
    x_col="file_paths",
    y_col="labels",
    target_size=img_size,
    color_mode=mode,
    class_mode='categorical',
    batch_size=batch_size,
    shuffle=False,
)
val_data = val_datagen.flow_from_dataframe(
    val_df,
    x_col="file_paths",
    y_col="labels",
    target_size=img_size,
    color_mode=mode,
    class_mode='categorical',
    batch_size=batch_size,
    shuffle=True,
)

# Get class names
class_names = list(train_data.class_indices.keys())
num_classes = len(class_names)

# Model configurations
MODEL_CONFIG = {
    'VGG16': {
        'model_class': tf.keras.applications.VGG16,
        'preprocess_input': tf.keras.applications.vgg16.preprocess_input,
        'layers_to_unfreeze': 4,
        'input_shape': (224, 224, 3)
    },
    'MobileNetV2': {
        'model_class': tf.keras.applications.MobileNetV2,
        'preprocess_input': tf.keras.applications.mobilenet_v2.preprocess_input,
        'layers_to_unfreeze': 20,
        'input_shape': (224, 224, 3)
    },
    'DenseNet121': {
        'model_class': tf.keras.applications.DenseNet121,
        'preprocess_input': tf.keras.applications.densenet.preprocess_input,
        'layers_to_unfreeze': 15,
        'input_shape': (224, 224, 3)
    },
    'ResNet101': {
        'model_class': tf.keras.applications.ResNet101,
        'preprocess_input': tf.keras.applications.resnet.preprocess_input,
        'layers_to_unfreeze': 30,
        'input_shape': (224, 224, 3)
    }
}

# Function to create model
def create_model(model_name):
    config = MODEL_CONFIG[model_name]
    
    # Create base model
    base_model = config['model_class'](
        weights='imagenet',
        include_top=False,
        input_shape=config['input_shape']
    )
    
    # Freeze base model
    base_model.trainable = False
    
    # Create model
    model = Sequential([
        Input(shape=config['input_shape']),
        base_model,
        GlobalAveragePooling2D(),
        Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.016)),
        Dropout(0.5),
        Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.016)),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer=Adamax(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model, base_model

# Function to train model
def train_model(model, model_name, epochs=25):
    # Create callbacks
    checkpoint = ModelCheckpoint(
        f'{model_name}_model.keras',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    )
    early_stopping = EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        mode='max'
    )
    
    # Train model
    history = model.fit(
        train_data,
        validation_data=val_data,
        epochs=epochs,
        callbacks=[checkpoint, early_stopping]
    )
    
    return history

# Function to evaluate model
def evaluate_model(model, model_name):
    # Evaluate on test data
    test_loss, test_acc = model.evaluate(test_data)
    print(f"{model_name} Test Accuracy: {test_acc:.4f}")
    
    # Make predictions
    predictions = model.predict(test_data)
    y_pred = np.argmax(predictions, axis=1)
    y_true = test_data.classes
    
    # Print classification report
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))
    
    # Plot confusion matrix
    plt.figure(figsize=(12, 9))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Blues)
    plt.xlabel('Predicted')
    plt.ylabel('Truth')
    plt.title(f"{model_name} Confusion Matrix")
    plt.show()
    
    return test_acc, y_pred, y_true

# Function to plot training history
def plot_history(history, model_name):
    # Create dataframe from history
    vals = pd.DataFrame(history.history)
    vals = vals.rename(columns={
        "loss": "Train Loss",
        "val_loss": "Validation Loss",
        "accuracy": "Train Accuracy",
        "val_accuracy": "Validation Accuracy"
    })
    
    # Plot accuracy
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(vals["Train Accuracy"], label="Training Accuracy")
    plt.plot(vals["Validation Accuracy"], label="Validation Accuracy")
    plt.title(f"{model_name} Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(vals["Train Loss"], label="Training Loss")
    plt.plot(vals["Validation Loss"], label="Validation Loss")
    plt.title(f"{model_name} Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Function to fine-tune model
def fine_tune_model(model, base_model, model_name, config):
    # Unfreeze top layers
    base_model.trainable = True
    
    # Freeze all layers except the last N layers
    for layer in base_model.layers[:-config['layers_to_unfreeze']]:
        layer.trainable = False
    
    # Re-compile with lower learning rate
    model.compile(
        optimizer=Adam(learning_rate=1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Create callbacks
    checkpoint = ModelCheckpoint(
        f'{model_name}_fine_tuned.keras',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    )
    early_stopping = EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        mode='max'
    )
    
    # Fine-tune model
    history = model.fit(
        train_data,
        validation_data=val_data,
        epochs=100,
        callbacks=[checkpoint, early_stopping]
    )
    
    return history

# Train and evaluate models
results = {}
models_to_train = ['VGG16', 'MobileNetV2', 'DenseNet121', 'ResNet101']

for model_name in models_to_train:
    print(f"\n{'='*50}")
    print(f"Training {model_name}")
    print(f"{'='*50}")
    
    # Create model
    model, base_model = create_model(model_name)
    model.summary()
    
    # Train model
    history = train_model(model, model_name, epochs)
    
    # Plot training history
    plot_history(history, model_name)
    
    # Evaluate model
    test_acc, y_pred, y_true = evaluate_model(model, model_name)
    
    # Fine-tune model
    print(f"\nFine-tuning {model_name}")
    fine_tune_history = fine_tune_model(model, base_model, model_name, MODEL_CONFIG[model_name])
    
    # Plot fine-tuning history
    plot_history(fine_tune_history, f"{model_name} (Fine-tuned)")
    
    # Evaluate fine-tuned model
    fine_tune_test_acc, _, _ = evaluate_model(model, f"{model_name} (Fine-tuned)")
    
    # Store results
    results[model_name] = {
        'initial_accuracy': test_acc,
        'fine_tuned_accuracy': fine_tune_test_acc
    }

# Compare models
print("\n" + "="*50)
print("Model Comparison")
print("="*50)

for model_name, result in results.items():
    print(f"{model_name}:")
    print(f"  Initial Accuracy: {result['initial_accuracy']:.4f}")
    print(f"  Fine-tuned Accuracy: {result['fine_tuned_accuracy']:.4f}")
    print(f"  Improvement: {result['fine_tuned_accuracy'] - result['initial_accuracy']:.4f}")

# Plot model comparison
model_names = list(results.keys())
initial_accs = [results[name]['initial_accuracy'] for name in model_names]
fine_tuned_accs = [results[name]['fine_tuned_accuracy'] for name in model_names]

plt.figure(figsize=(12, 6))
x = np.arange(len(model_names))
width = 0.35

plt.bar(x - width/2, initial_accs, width, label='Initial Accuracy')
plt.bar(x + width/2, fine_tuned_accs, width, label='Fine-tuned Accuracy')

plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Comparison')
plt.xticks(x, model_names)
plt.legend()
plt.ylim(0, 1)
plt.show()
```
