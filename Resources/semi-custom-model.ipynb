```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (confusion_matrix, classification_report,
                             precision_score, recall_score, f1_score,
                             ConfusionMatrixDisplay)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import (GlobalAveragePooling2D, Dense, Dropout,
                                     BatchNormalization, Conv2D, MaxPooling2D,
                                     Input, Activation)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, Callback
from tensorflow.keras.regularizers import l2

# Set seeds
np.random.seed(42)
tf.random.set_seed(42)

print("="*70)
print("üöÄ FIXED SEMI-SUPERVISED LEARNING FOR MANGO LEAF DISEASE")
print("="*70)

# ============================================================================
# CONFIGURATION
# ============================================================================
DATA_DIR = '/kaggle/input/mango-leaf-disease-dataset'
IMG_SIZE = (224, 224)
BATCH_SIZE = 32  # Smaller batch = better generalization with small dataset
SEED = 42
LABELED_FRACTION = 0.25

# Semi-supervised configuration
SSL_CONFIG = {
    'num_iterations': 4,  # More iterations with smaller steps
    'initial_threshold': 0.70,  # Start very conservative
    'final_threshold': 0.90,  # More conservative final threshold
    'min_samples_per_iteration': 50,
    'max_samples_per_iteration': 300,  # Don't add too many at once
    'initial_epochs': 50,  # More epochs with early stopping
    'finetune_epochs': 20,  # Fewer epochs per iteration
}

# ============================================================================
# LOAD DATA
# ============================================================================
print("\n=== Loading Dataset ===")
filepaths = []
labels = []
for label in sorted(os.listdir(DATA_DIR)):
    label_path = os.path.join(DATA_DIR, label)
    if os.path.isdir(label_path):
        for file in os.listdir(label_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                filepaths.append(os.path.join(label_path, file))
                labels.append(label)

df = pd.DataFrame({'filename': filepaths, 'class': labels})
classes = sorted(df['class'].unique())
num_classes = len(classes)

print(f"Total samples: {len(df)}")
print(f"Classes: {classes}")

# ============================================================================
# SPLIT DATA
# ============================================================================
train_val_df, test_df = train_test_split(
    df, test_size=0.1, stratify=df['class'], random_state=SEED
)
train_df, val_df = train_test_split(
    train_val_df, test_size=0.1111, stratify=train_val_df['class'], random_state=SEED
)
labeled_df, unlabeled_df = train_test_split(
    train_df, test_size=(1 - LABELED_FRACTION), 
    stratify=train_df['class'], random_state=SEED
)

print(f"\nData split:")
print(f"  Labeled: {len(labeled_df)}")
print(f"  Unlabeled: {len(unlabeled_df)}")
print(f"  Validation: {len(val_df)}")
print(f"  Test: {len(test_df)}")

# ============================================================================
# DATA GENERATORS WITH AGGRESSIVE AUGMENTATION
# ============================================================================
print("\n=== Creating Data Generators ===")

# CRITICAL: Aggressive augmentation to prevent overfitting on small dataset
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.25,
    height_shift_range=0.25,
    shear_range=0.2,
    zoom_range=0.25,
    horizontal_flip=True,
    vertical_flip=True,  # Leaves can be upside down
    brightness_range=[0.7, 1.3],
    fill_mode='nearest',
    # Advanced augmentations
    channel_shift_range=20.0,
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

class_names = sorted(df['class'].unique())

# ============================================================================
# BUILD REGULARIZED MODEL
# ============================================================================
def build_model(input_shape=(224, 224, 3), num_classes=8):
    """
    Model with heavy regularization to prevent overfitting
    """
    inputs = Input(shape=input_shape)
    
    # Block 1
    x = Conv2D(32, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.2)(x)
    
    # Block 2
    x = Conv2D(64, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.3)(x)
    
    # Block 3
    x = Conv2D(128, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.4)(x)
    
    # Block 4
    x = Conv2D(256, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.5)(x)
    
    # Head with heavy dropout
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.6)(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

print("\n=== Building Model ===")
model = build_model(input_shape=IMG_SIZE + (3,), num_classes=num_classes)
print(f"Model parameters: {model.count_params():,}")

# ============================================================================
# CUSTOM CALLBACK
# ============================================================================
class TrainingMonitor(Callback):
    def __init__(self, phase_name):
        super().__init__()
        self.phase_name = phase_name
        self.best_val_acc = 0
        
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        val_acc = logs.get('val_accuracy', 0)
        if val_acc > self.best_val_acc:
            self.best_val_acc = val_acc
            print(f"\n[{self.phase_name}] üéØ New best val_acc: {val_acc:.4f}")

# ============================================================================
# PHASE 1: INITIAL SUPERVISED TRAINING
# ============================================================================
print("\n" + "="*70)
print("PHASE 1: INITIAL SUPERVISED TRAINING")
print("="*70)

labeled_gen = train_datagen.flow_from_dataframe(
    dataframe=labeled_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True,
    seed=SEED,
    classes=class_names
)

val_gen = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    classes=class_names
)

test_gen = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col='filename',
    y_col='class',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False,
    classes=class_names
)

# Compile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=5e-4),  # Lower LR to prevent overfitting
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Aggressive early stopping
early_stop_phase1 = EarlyStopping(
    monitor='val_accuracy',  # Monitor val accuracy, not loss
    patience=15,
    restore_best_weights=True,
    mode='max',
    verbose=1
)

reduce_lr_phase1 = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.5,
    patience=7,
    min_lr=1e-6,
    mode='max',
    verbose=1
)

monitor_phase1 = TrainingMonitor("PHASE 1")

# Train
print("\nTraining with heavy augmentation and regularization...")
history_phase1 = model.fit(
    labeled_gen,
    validation_data=val_gen,
    epochs=SSL_CONFIG['initial_epochs'],
    callbacks=[early_stop_phase1, reduce_lr_phase1, monitor_phase1],
    verbose=2
)

# Evaluate Phase 1
phase1_val_loss, phase1_val_acc = model.evaluate(val_gen, verbose=0)
print(f"\nüìä Phase 1 Results:")
print(f"   Val Accuracy: {phase1_val_acc:.4f}")
print(f"   Val Loss: {phase1_val_loss:.4f}")

if phase1_val_acc < 0.3:
    print("\n‚ö†Ô∏è  WARNING: Phase 1 validation accuracy is low!")
    print("   Consider training longer or adjusting hyperparameters")
    print("   Continuing with semi-supervised learning anyway...")

# ============================================================================
# PHASE 2: ITERATIVE PSEUDO-LABELING
# ============================================================================
print("\n" + "="*70)
print("PHASE 2: ITERATIVE PSEUDO-LABELING")
print("="*70)

all_histories = [history_phase1]
current_unlabeled_df = unlabeled_df.copy()
current_labeled_df = labeled_df.copy()

for iteration in range(SSL_CONFIG['num_iterations']):
    print(f"\n{'='*70}")
    print(f"ITERATION {iteration + 1}/{SSL_CONFIG['num_iterations']}")
    print(f"{'='*70}")
    
    # Progressive confidence threshold
    progress = iteration / max(SSL_CONFIG['num_iterations'] - 1, 1)
    threshold = SSL_CONFIG['initial_threshold'] + \
                (SSL_CONFIG['final_threshold'] - SSL_CONFIG['initial_threshold']) * progress
    
    print(f"Confidence threshold: {threshold:.3f}")
    print(f"Remaining unlabeled: {len(current_unlabeled_df)}")
    
    if len(current_unlabeled_df) == 0:
        print("No unlabeled samples remaining.")
        break
    
    # Generate pseudo-labels
    print("\nGenerating pseudo-labels...")
    unlabeled_gen = val_test_datagen.flow_from_dataframe(
        dataframe=current_unlabeled_df[['filename']],
        x_col='filename',
        y_col=None,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode=None,
        shuffle=False
    )
    
    # Use test-time augmentation for more robust predictions
    n_tta = 3  # Test-time augmentation rounds
    all_preds = []
    
    for tta_round in range(n_tta):
        preds = model.predict(unlabeled_gen, verbose=0)
        all_preds.append(preds)
    
    # Average predictions across TTA
    unlabeled_preds = np.mean(all_preds, axis=0)
    
    predicted_class_indices = np.argmax(unlabeled_preds, axis=1)
    max_probs = np.max(unlabeled_preds, axis=1)
    
    # Filter by confidence
    confident_mask = max_probs >= threshold
    num_confident = confident_mask.sum()
    
    print(f"High-confidence samples: {num_confident} ({100*num_confident/len(current_unlabeled_df):.1f}%)")
    print(f"Confidence: min={max_probs.min():.3f}, mean={max_probs.mean():.3f}, max={max_probs.max():.3f}")
    
    if num_confident < SSL_CONFIG['min_samples_per_iteration']:
        print(f"Too few samples ({num_confident} < {SSL_CONFIG['min_samples_per_iteration']}), skipping")
        continue
    
    # Limit samples to prevent sudden distribution shift
    if num_confident > SSL_CONFIG['max_samples_per_iteration']:
        confident_indices = np.where(confident_mask)[0]
        # Select highest confidence samples
        top_indices = confident_indices[np.argsort(-max_probs[confident_mask])[:SSL_CONFIG['max_samples_per_iteration']]]
        print(f"Limiting to top {SSL_CONFIG['max_samples_per_iteration']} most confident samples")
    else:
        top_indices = np.where(confident_mask)[0]
    
    # Class distribution
    selected_classes = predicted_class_indices[top_indices]
    print("\nPseudo-label distribution:")
    for cls_idx in range(num_classes):
        count = (selected_classes == cls_idx).sum()
        if count > 0:
            print(f"  {classes[cls_idx]}: {count} ({100*count/len(top_indices):.1f}%)")
    
    # Create pseudo-labeled dataset
    pseudo_filenames = current_unlabeled_df.iloc[top_indices]['filename'].values
    pseudo_classes = [classes[idx] for idx in predicted_class_indices[top_indices]]
    
    pseudo_df = pd.DataFrame({
        'filename': pseudo_filenames,
        'class': pseudo_classes
    })
    
    # Combine with labeled data
    current_labeled_df = pd.concat([current_labeled_df, pseudo_df], ignore_index=True)
    print(f"\nTraining set: {len(labeled_df)} original + {len(current_labeled_df) - len(labeled_df)} pseudo = {len(current_labeled_df)} total")
    
    # Create combined generator
    combined_gen = train_datagen.flow_from_dataframe(
        dataframe=current_labeled_df,
        x_col='filename',
        y_col='class',
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True,
        seed=SEED,
        classes=class_names
    )
    
    # Fine-tune with even lower learning rate
    current_lr = 1e-4 / (1.5 ** iteration)
    model.compile(
        optimizer=Adam(learning_rate=current_lr),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    print(f"Learning rate: {current_lr:.2e}")
    
    # Callbacks
    early_stop_iter = EarlyStopping(
        monitor='val_accuracy',
        patience=8,
        restore_best_weights=True,
        mode='max',
        verbose=1
    )
    
    reduce_lr_iter = ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=4,
        min_lr=1e-7,
        mode='max',
        verbose=1
    )
    
    monitor_iter = TrainingMonitor(f"ITER {iteration+1}")
    
    # Train
    history_iter = model.fit(
        combined_gen,
        validation_data=val_gen,
        epochs=SSL_CONFIG['finetune_epochs'],
        callbacks=[early_stop_iter, reduce_lr_iter, monitor_iter],
        verbose=2
    )
    
    all_histories.append(history_iter)
    
    # Evaluate
    iter_val_loss, iter_val_acc = model.evaluate(val_gen, verbose=0)
    improvement = iter_val_acc - phase1_val_acc
    print(f"\nüìä Iteration {iteration+1} Results:")
    print(f"   Val Accuracy: {iter_val_acc:.4f} ({'+' if improvement > 0 else ''}{improvement:.4f})")
    
    # Remove used samples
    current_unlabeled_df = current_unlabeled_df.drop(
        current_unlabeled_df.index[top_indices]
    ).reset_index(drop=True)

# ============================================================================
# FINAL EVALUATION
# ============================================================================
print("\n" + "="*70)
print("FINAL EVALUATION")
print("="*70)

test_loss, test_acc = model.evaluate(test_gen, verbose=1)
print(f"\nüéØ Test Accuracy: {test_acc:.4f}")
print(f"üìâ Test Loss: {test_loss:.4f}")
print(f"üìà Improvement: {test_acc - phase1_val_acc:+.4f} ({100*(test_acc - phase1_val_acc):+.1f}%)")

# Detailed metrics
y_pred_probs = model.predict(test_gen, verbose=0)
y_pred_labels = np.argmax(y_pred_probs, axis=1)
y_true_labels = test_gen.classes

print("\n" + "="*70)
print("CLASSIFICATION REPORT")
print("="*70)
print(classification_report(y_true_labels, y_pred_labels, target_names=classes))

# Metrics
test_precision = precision_score(y_true_labels, y_pred_labels, average='macro')
test_recall = recall_score(y_true_labels, y_pred_labels, average='macro')
test_f1 = f1_score(y_true_labels, y_pred_labels, average='macro')

print(f"\nüìä Summary:")
print(f"   Precision: {test_precision:.4f}")
print(f"   Recall: {test_recall:.4f}")
print(f"   F1-Score: {test_f1:.4f}")

# ============================================================================
# VISUALIZATION
# ============================================================================
def combine_histories(histories):
    combined = {'accuracy': [], 'val_accuracy': [], 'loss': [], 'val_loss': []}
    for hist in histories:
        for key in combined.keys():
            combined[key].extend(hist.history[key])
    return combined

combined_history = combine_histories(all_histories)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Accuracy
axes[0].plot(combined_history['accuracy'], label='Training', linewidth=2)
axes[0].plot(combined_history['val_accuracy'], label='Validation', linewidth=2)
axes[0].axvline(len(history_phase1.history['accuracy']), color='red', 
                linestyle='--', alpha=0.5, label='SSL Start')
axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Loss
axes[1].plot(combined_history['loss'], label='Training', linewidth=2)
axes[1].plot(combined_history['val_loss'], label='Validation', linewidth=2)
axes[1].axvline(len(history_phase1.history['loss']), color='red', 
                linestyle='--', alpha=0.5, label='SSL Start')
axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
plt.show()

# Confusion Matrix
cm = confusion_matrix(y_true_labels, y_pred_labels)
plt.figure(figsize=(12, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot(cmap='Blues', values_format='d', xticks_rotation=45)
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n‚úÖ Semi-supervised learning completed!")
```
